<h1 style="color:brown">Linear Regression</h1>

<h3>Linear Regression is a basic statistical technique used to generalize the given data using a line equation. A linear regressor accepts a series of values mapping dimensions x and y, and calculates parameters 'm' (slope) and 'c' (bias), consitituting the line equation <br></h3><h3 style="text-align:center"><i> <b>y</b>= m<b>x</b> + c</i></h3> <h3><br>  The line equation produced through linear regression can be used to approximate the value of 'y' from the data for any given value of 'x'.</h3>

<h1 style="color:brown">Implementation</h1>

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing data
train_data = pd.read_csv("src/train.csv")
test_data = pd.read_csv("src/test.csv")
#data sourced from https://www.kaggle.com/datasets/andonians/random-linear-regression



print(train_data)
```

```python
#cleaning data

print(train_data.isna().any())
train_data.dropna(inplace = True)
```

```python
#instantiating the LinearRegressor class
myRegressor = LinearRegression()

#splitting feature and target variables of the train data. Feature variables are stored in X whereas the target feature is stored in y.
trainX = pd.Series(train_data.x).values.reshape(-1,1)
trainY = pd.Series(train_data.y).values.reshape(-1,1)

#fitting the regressor to the given train data
myRegressor.fit(trainX, trainY)


#using the regressor to predict target variable of the test data data
predicted = myRegressor.predict(test_data.x.values.reshape(-1,1))
print(predicted)
```

```python
test_data.y.shape
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(20,10))
plt.xlabel("test_data.x")
plt.ylabel("target feature")

#Comparing target feature's real values from test data and the values predicted by our linear regression model 

plt.scatter(test_data.x,test_data.y, color="skyblue", label= "real values")
plt.plot(test_data.x, predicted, color = "purple", label= "predicted")

plt.legend()
```

```python
#Calculating error

print("Error between the predicted and the real values: ", mean_absolute_error(predicted.reshape(-1,), test_data.y))
```
