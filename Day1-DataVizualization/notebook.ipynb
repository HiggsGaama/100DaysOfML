```python
import pandas as pd 

import seaborn as sns
import matplotlib.pyplot as plt


sns.set_style("darkgrid")
```

<h1 style="color:skyblue" >Part 1: Listing various vizualization methods</h1>

```python
#Loading the dataset
data = sns.load_dataset("iris")

print(data)
```

<h3>1.Scatter Plot</h3>
<h5>Scatter plots are used to find the relation between bi-variate data</h5>

```python
plt.figure(figsize=(15,7))

plt.title("Scatter of sepal_length")
plt.xlabel("sepal_length")
plt.ylabel("Range")


#using seaborn
sns.scatterplot( x=data.sepal_length, y = sorted(data.sepal_length))  

#Using matplotlib
#plt.scatter(x=data.sepal_length, y = sorted(data.sepal_length))  
```

<h2>1.5 Hexbin Plot</h2>
<h5>Hexbins improve the visualization of scatterplots by color encoding the points based on density</h5>

```python
plt.figure(figsize=(15,7))

plt.title("Scatter of sepal_length")
plt.xlabel("sepal_length")
plt.ylabel("Range")

plt.hexbin(x=data.sepal_length, y = sorted(data.sepal_length))
```

<h3>2.Histogram</h3>
<h5>Histograms help in understanding the distrubution of data about any one particular dimension</h5>

```python
plt.figure(figsize=(15,7))

plt.title("Distribution of sepal_length")
plt.xlabel("sepal_lenth")
plt.ylabel("frequency")

#using seaborn
sns.histplot(data = data, x = "sepal_length", bins = 30, kde = True)  

#Using matplotlib
#plt.hist(data.sepal_length, bins = 30)     
```

<h3>3. Bar Graphs</h3>
<h5>Bar graphs help visualize the frequency or an estimate of central tendency of different groups in a single dimenion of data.</h5>

```python
plt.figure(figsize=(15,7))


#Using matplotlib
#plt.bar(data.species, data.sepal_length) #Using matplotlib

#using seaborn
sns.barplot(data = data ,x= "species", y = "sepal_length") 


print(data.species.value_counts())
```

<h3>4. Pie Chart</h3>
<h5> Pie charts help visualize proportions of each category in categorical data</h5>

```python
plt.figure(figsize=(10,5))

#Using matplotlib
plt.pie(data.species.value_counts().values, labels = data.species.value_counts().keys()   ,explode= [0.2, 0.05 , 0.05]) 

#seaborn has no function to make piecharts
```

<h2>5.Box plot</h2>
<h5> Box plots help understand the distribution of data using the five-number summary method that displays the values - Minimum, 1st Quartile, Median, 3rd Quartile, Maximum. </h5>

```python
plt.figure(figsize=(15, 7))

plt.title("Box plot of numerical dimensions of the iris dataset")
plt.ylabel("Distribution")

#using matplotlib
#plt.boxplot([data.sepal_length, data.sepal_width, data.petal_length, data.petal_width], labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']) 

#using seaborn
sns.boxplot(data = data)    
```

<h2>5. Violin plot</h2>
<h5> Violin plots are box plots with an additional rotated Kernel Density Estimates</h5>

```python
plt.figure(figsize=(15, 7))

plt.title("Violin plot of numerical dimensions of the iris dataset")
plt.ylabel("Distribution")

#using matplotlib
#plt.violinplot([data.sepal_length, data.sepal_width, data.petal_length, data.petal_width])

#using seaborn
sns.violinplot(data = data)
```

<h2> 6. Heatmap</h2>
<h5>Heatmaps are a type of matrix plots that allow you to plot matrix data as color-encoded matrices. They are mostly used to find multi-collinearity in a dataset.</h5>

```python
plt.figure(figsize=(15,7))

plt.title("Heatmap representing correlations between numerical entities in Iris dataset")

print("Correlation matrix :\n", data.corr())

#using seaborn
sns.heatmap(data.corr(), cmap = "crest", annot = True)

#matplotlib has no functions to make heatmaps
```
